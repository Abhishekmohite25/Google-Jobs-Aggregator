```python
import matplotlib.pyplot as plt
```

### Initializing Spark Session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Visualization").getOrCreate()
```

```python
df = spark.read.table("lakehouse.gold.jobs")

#display(df)

pdf = df.orderBy("posted_at", ascending=True).limit(20).toPandas()

#display(pdf)
```

### Top Companies Hiring

```python
plt.figure(figsize=(10,5))
plt.barh(pdf['company_name'][::-1], pdf['search_role'][::-1], color='skyblue')
plt.xlabel('Number of Jobs')
plt.ylabel('Top Companies Hiring')
plt.xticks(rotation=90, ha='center')
plt.grid(True,axis='x',linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```

### Job Demand by City

```python
df_cities = spark.read.table("lakehouse.gold.jobs")
pdf_city = df_cities.orderBy("posted_at", ascending=True).toPandas()

# Plot
plt.figure(figsize=(10, 6))
plt.barh(pdf_city['location'][::-1], pdf_city['title'][::-1], color='lightgreen')
plt.xlabel("Job Postings")
plt.title("Job Demand by City")
plt.xticks(rotation=90, ha='center')
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
